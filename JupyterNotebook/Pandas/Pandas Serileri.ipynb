# Pandas Serileri

```python
import numpy as np
import pandas as pd
```

# Tanımlanma Yöntemleri - 1

```python
labels_list = ["Taner","Mesis","Ronlado","Samir","Zeynep"]
```

```python
data_list = [10,20,30,40,50]
```

```python
pd.Series(data = data_list)
```

```python
pd.Series(data = data_list,index = labels_list)
```


# Tanımlanma Yöntemleri - 2
