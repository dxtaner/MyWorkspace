# Array Operasyonları


```python
arr1 = ((np.random.rand(9) * 5).astype(int)).reshape(3,3)
arr2 = ((np.random.rand(9) * 5).astype(int)).reshape(3,3)
```

```python
arr1
```

```python
arr2
```

# İki Matrisi Toplayın

```python
result = np.add(arr1, arr2)
```

```python
result
```

# İki Matrisi Çarpın

```python
result = np.dot(arr1, arr2)
```

```python
result
```
