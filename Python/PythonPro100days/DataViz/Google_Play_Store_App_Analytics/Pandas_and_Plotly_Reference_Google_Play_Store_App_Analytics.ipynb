```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("apps.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df = df[df["Rating"] <= 5]
```

```python
df["Installs"] = (
    df["Installs"]
    .str.replace(",", "")
    .str.replace("+", "")
    .astype(int)
)
```

```python
df["Price"] = df["Price"].str.replace("$", "").astype(float)
```

```python
category_count = df["Category"].value_counts()
category_count
```

```python
category_installs = (
    df.groupby("Category")["Installs"]
    .sum()
    .sort_values(ascending=False)
)
category_installs
```

```python
fig = px.bar(
    x=category_installs.index,
    y=category_installs.values,
    title="Total Installs by Category",
    labels={"x": "Category", "y": "Total Installs"}
)
fig.show()
```

```python
df["Type"].value_counts()
```

```python
fig = px.box(
    df,
    x="Type",
    y="Rating",
    title="App Ratings: Free vs Paid"
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="Reviews",
    y="Installs",
    color="Type",
    title="Reviews vs Installs"
)
fig.show()
```

```python
paid_apps = df[df["Type"] == "Paid"]

fig = px.scatter(
    paid_apps,
    x="Price",
    y="Rating",
    title="Price vs Rating (Paid Apps)"
)
fig.show()
```

```python
df.columns
```

```python
content_rating = df["Content_Rating"].value_counts()
content_rating
```

```python
fig = px.pie(
    values=content_rating.values,
    names=content_rating.index,
    title="Content Rating Distribution"
)
fig.show()
```

```python
df["Last_Updated"] = pd.to_datetime(df["Last_Updated"])

updates_per_year = df["Last_Updated"].dt.year.value_counts().sort_index()
updates_per_year
```

```python
fig = px.line(
    x=updates_per_year.index,
    y=updates_per_year.values,
    title="Number of App Updates per Year"
)
fig.show()
```
