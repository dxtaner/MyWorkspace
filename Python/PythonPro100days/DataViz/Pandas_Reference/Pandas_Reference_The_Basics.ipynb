```python
# Pandas kütüphanesini içe aktar (veri okuma, temizleme, analiz için)
import pandas as pd
```

```python
# Veri setinin dosya adı
file_path = "salaries_by_college_major.csv"
```

```python
# CSV dosyasını okumayı dene
try:
    df = pd.read_csv(file_path)     # CSV dosyasını okuma
    print("Dosya başarıyla yüklendi:", file_path)
except FileNotFoundError:
    print(f"Dosya bulunamadı: {file_path} — Lütfen dosya yolunu kontrol edin.")
```

```python
# Veri setinin başlangıç kısmını göster (kontrol için)
df.head()
```

```python
# Kolonlar, veri tipleri, kaç satır olduğu gibi bilgileri gösterir
df.info()
```

```python
# Ortalama, min, max gibi özet istatistikleri gösterme
df.describe()
```

```python
# Hangi kolonlarda kaç tane eksik değer olduğunu gösterir
df.isna().sum()
```

```python
# Eksik değer içeren satırları temizleme
df = df.dropna()
```

```python
# 'Undergraduate Major' kolonunun ilk 5 değerini göster
df["Undergraduate Major"].head()
```

```python
# Birden fazla kolonu görüntüleme
df[["Undergraduate Major", "Starting Median Salary"]].head()
```

```python
# 1. satırdaki Mid-Career Median Salary değerini getir
df.loc[1, "Mid-Career Median Salary"]
```

```python
# Starting Median Salary'e göre büyükten küçüğe sıralama
df.sort_values("Starting Median Salary", ascending=False).head(10)
```

```python
# Starting Median Salary'e göre küçükten büyüğe sıralama
df.sort_values("Starting Median Salary").head(10)
```

```python
# Potansiyel = Mid-Career 90th Percentile - Mid-Career 10th Percentile
df["Potential"] = df["Mid-Career 90th Percentile Salary"] - df["Mid-Career 10th Percentile Salary"]
```

```python
df.nlargest(10, "Potential")
```

```python
# Group (Business, STEM, Humanities vb.) kolonuna göre gruplayıp
# Mid-Career Median Salary ortalamasını alıyoruz
df.groupby("Group")["Mid-Career Median Salary"].mean()
```

```python
df.pivot_table(
    values="Mid-Career Median Salary",
    index="Group",
    aggfunc="mean"
)
```
