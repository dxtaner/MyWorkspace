```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("film_budget_revenue_dataset.csv")
```

```python
df["Release_Date"] = pd.to_datetime(df["Release_Date"])
df["Release_Year"] = df["Release_Date"].dt.year
```

```python
money_cols = [
    "USD_Production_Budget",
    "USD_Worldwide_Gross",
    "USD_Domestic_Gross"
]

for col in money_cols:
    df[col] = (
        df[col]
        .replace('[\$,]', '', regex=True)
        .astype(float)
    )
```

```python
df = df[df["USD_Production_Budget"] > 0]
df = df[df["USD_Worldwide_Gross"] > 0]
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df,
    alpha=0.6
)
plt.title("Budget vs Worldwide Gross")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df
)
plt.title("Linear Regression (Raw Scale)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df,
    scatter_kws={"alpha": 0.5}
)
```

```python
plt.figure(figsize=(8,5))
sns.regplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df,
    scatter_kws={"alpha": 0.5}
)
```

```python
sns.regplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df,
    ci=None
)
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
sns.regplot(
    x="budget",
    y="revenue",
    data=df
)
```

```python
sns.regplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df,
    scatter=False
)
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
sns.regplot(
    x="USD_Production_Budget",
    y="USD_Worldwide_Gross",
    data=df,
    order=2
)
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
sns.regplot(
    x="USD_Domestic_Gross",
    y="USD_Worldwide_Gross",
    data=df
)
plt.xscale("log")
plt.yscale("log")
plt.title("Domestic vs Worldwide Gross")
plt.show()
```
