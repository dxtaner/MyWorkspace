```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("fatal-police-shootings-data.csv")
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"])
df["year"] = df["date"].dt.year
```

```python
yearly_deaths = df["year"].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(yearly_deaths.index, yearly_deaths.values)
plt.title("Deaths Involving Police by Year")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.show()
```

```python
race_counts = df["race"].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=race_counts.index, y=race_counts.values)
plt.title("Deaths by Race")
plt.xlabel("Race")
plt.ylabel("Count")
plt.show()
```

```python
gender_counts = df["gender"].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title("Deaths by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
mental_counts = df["signs_of_mental_illness"].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=mental_counts.index, y=mental_counts.values)
plt.title("Mental Illness Indicator")
plt.xlabel("Mental Illness (True / False)")
plt.ylabel("Count")
plt.show()
```

```python
state_counts = df["state"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=state_counts.index, y=state_counts.values)
plt.title("Top 10 States by Police-Involved Deaths")
plt.xlabel("State")
plt.ylabel("Deaths")
plt.show()
```
