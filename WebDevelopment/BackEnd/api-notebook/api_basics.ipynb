API (Application Programming Interface), istemci ile sunucu arasında veri alışverişini sağlar.

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route("/")
def home():
    return jsonify({"message": "API çalışıyor"})

@app.route("/users")
def users():
    return jsonify([
        {"id": 1, "name": "Taner"},
        {"id": 2, "name": "Ahmet"}
    ])

app.run(debug=True)
```

```python
import requests

response = requests.get("http://127.0.0.1:5000/users")
response.json()
```

## HTTP Metodları
- GET → Veri çekme
- POST → Veri ekleme
- PUT → Veri güncelleme
- DELETE → Veri silme
